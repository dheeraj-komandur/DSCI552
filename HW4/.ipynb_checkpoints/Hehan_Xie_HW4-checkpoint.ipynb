#### (i) Fit a PCR model on the training set, with M (the number of principal components) chosen by cross-validation. Report the test error obtained.

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_validate

bestM = 1
bestMSE = 1
M = range(1, 123)
mse = []
for m in M:
    # get Principal Components
    pca = PCA(n_components=m)
    X_reduced = pca.fit_transform(xTrain)
    # do one CV to get MSE
    linearRegModel = LinearRegression()
    scores = cross_validate(linearRegModel, X_reduced, yTrain, cv=3,scoring=('r2', 'neg_mean_squared_error'))
#     print(scores)
    rSquare = np.mean(scores['test_r2'])
    mse = -1*(np.mean(scores['test_neg_mean_squared_error']))
    if mse < bestMSE:
        bestMSE = mse
        bestM = m
    print('M:', m)
    print('R-Square:', rSquare)
    print('MSE:', mse)
    print()
print('Best M:', bestM)
print('Best MSE:', bestMSE)
```

```python
# choose M = 86 to get Principal Components
m = bestM
pca = PCA(n_components=m)
X_reduced = pca.fit_transform(xTrain)
# do one CV to get MSE
scores = cross_validate(LinearRegression(), X_reduced, yTrain, cv=3,scoring=('r2', 'neg_mean_squared_error'))
rSquare = np.mean(scores['test_r2'])
mse = -1*(np.mean(scores['test_neg_mean_squared_error']))
print('Best Result:')
print('M:', m)
print('R-Square:', rSquare)
print('MSE:', mse)
```

#### (j) In this section, we would like to fit a boosting tree to the data. As in classification trees, one can use any type of regression at each node to build a multivariate regression tree. Because the number of variables is large in this problem, one can use L1-penalized regression at each node. Such a tree is called L1 penalized gradient boosting tree. You can use XGBoost3 to fit the model tree. Determine α (the regularization term) using cross-validation.

```python
import xgboost as xgb
import warnings
warnings.filterwarnings('ignore')
# convert to xgb matrix
trainMatrix = xgb.DMatrix(data=xTrain, label=yTrain)
aList = [1e-3, 1e-2, 1e-1, 1]
aList = np.power(10, aList)
for a in aList:
    param = {'max_depth':2, 'eta':0.3, 'silent':1, 'objective':'reg:linear', 'alpha':a}
    scores = xgb.cv(param, trainMatrix, num_boost_round=100, nfold=5,metrics='rmse')
    mse = np.mean(scores['test-rmse-mean'])
    print('α:', a)
    print('MSE:', mse)
```

```python
# the best result is
# α: 1.023292992280754
# MSE: 0.146043806
```

#### (c) Train a random forest to classify the data set. Do NOT compensate for class imbalance in the data set. Calculate the confusion matrix, ROC, AUC, and misclassification for training and test sets and report them (You may use pROC package). Calculate Out of Bag error estimate for your random forset and compare it to the test error.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.metrics import roc_auc_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc

# convert Neg/Pos to number, Negative:0;Positive:1
newYTrain = yTrain.apply(lambda x: 0 if x=='neg' else 1)
newYTest = yTest.apply(lambda x: 0 if x=='neg' else 1)
# print(newYTrain)
# build Random Forest Classifier
clf = RandomForestClassifier(n_estimators=100, max_depth=2,oob_score=True)
clf.fit(xTrain, newYTrain)
yTrainPredict = clf.predict(xTrain)
# print(yTrainPredict)

# draw ROC and AUC
yScore = clf.predict_proba(xTrain)[:, 1]
# Compute ROC curve and ROC area for each class
fpr,tpr,threshold = roc_curve(newYTrain, yScore)
# cal AUC value
roc_auc = auc(fpr,tpr)
# plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Train Data Random Forest ROC')
plt.legend(loc="lower right")
plt.show()

# calculate confusion matrix
confusionMatrix = confusion_matrix(newYTrain, yTrainPredict)
print('Train Data Confusion Matrix:')
print(confusionMatrix)
# calculate f1-score
print('F1-Score:', f1_score(newYTrain, yTrainPredict))
# calculate misclassification rate
print('Misclassification rate is:', 1-clf.score(xTrain, newYTrain))
# calculate out of bag error
print('Out of Bag error is:', 1-clf.oob_score_)
```

```python
yTestPredict = clf.predict(xTest)
# for Test Data


# draw ROC and AUC
yScore = clf.predict_proba(xTest)[:, 1]
# Compute ROC curve and ROC area for each class
fpr,tpr,threshold = roc_curve(newYTest, yScore)
# cal AUC value
roc_auc = auc(fpr,tpr)
# plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Test Data Random Forest ROC')
plt.legend(loc="lower right")
plt.show()

# calculate confusion matrix
confusionMatrix = confusion_matrix(newYTest, yTestPredict)
print('Test Data Confusion Matrix:')
print(confusionMatrix)
# calculate f1-score
print('F1-Score:', f1_score(newYTest, yTestPredict))
# calculate misclassification rate
print('Misclassification rate is:', 1-clf.score(xTest, newYTest))
# calculate out of bag error
print('Out of Bag error is:', 1-clf.oob_score_)
```

### 3. ISLR 6.8.3

(a) iv. Steadily decrease. 
    When s increase, the penalty will increate, which result in the model will be more flexible so that RSS will steadily decrease for training set.

(b) ii. Decrease initially, and then eventually start increasing in a U shape.
    When s increase, the model become more flexible, provokes at first a decrease in test RSS before increasing again after that in a typical U shape.

(c) iii. Steadily increase. 
    When s increase, the model become more flexible, which result in a steady increase in variance.

(d) iv. Steadily decrease. 
    When s increase, the penalty will increate, which result in the model will be more flexible so that bias will decrease.

(e) v.Remain constant. Because irreducible error is independant of the model, it is irrelevant with s.

### 4. ISLR, 6.8.5

(a) According to x11 = x12 = x1, x21 = x22 = 2, we can get this formula:

```python
from IPython.display import Latex
```

```python
Latex("$(y_1-β̂_1*x_1-β̂_2*x_1)^2 + (y_2-β̂_1*x_2-β̂_2*x_2)^2 + λ(β̂^2_1 + β̂^2_2)$")
```

(b) Set them equal to 0, we can obtain:

```python
Latex(r"$β̂_1(x^2_1 + x^2_2 + λ) + β̂_2(x^2_1 + x^2_2) = y_1x_1 + y_2x_2\\β̂_1(x^2_1 + x^2_2) + β̂_2(x^2_1 + x^2_2 + λ) = y_1x_1 + y_2x_2$")
```

We can get β̂1 = β̂2 from above.

(c) According to x11=x12=x1, x21=x22=x2, we need to minimize:

```python
Latex("$(y_1-β̂_1*x_1-β̂_2*x_1)^2 + (y_2-β̂_1*x_2-β̂_2*x_2)^2 + λ(|β̂_1|+|β̂_2|)$")
```

(d) Because x1+x2=0 and y1+y2=0, we can simplify (c) formula as:

```python
Latex("$2(y_1-(β̂_1+β̂_2)*x_1)^2 >=0 \\\β̂_1+β̂_2 = y_1/x_1 \\\ β̂_1+β̂_2 = s \; or\; β̂_1+β̂_2 = -s$")
```
